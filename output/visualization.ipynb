```python
from math import atan2
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import HoverTool
from bokeh.models import ColumnDataSource

# Read the data from the file
data = []
with open("output.txt", "r") as file:
    for line in file:
        values = line.strip().split()
        data.append(values)

# Create empty ColumnDataSources for each sensor type
radar_source = ColumnDataSource(data=dict(
    est_px=[], est_py=[], est_vx=[], est_vy=[],
    meas_px=[], meas_py=[],
    gt_px=[], gt_py=[], gt_vx=[], gt_vy=[]
))

lidar_source = ColumnDataSource(data=dict(
    est_px=[], est_py=[], est_vx=[], est_vy=[],
    meas_px=[], meas_py=[],
    gt_px=[], gt_py=[], gt_vx=[], gt_vy=[]
))

combined_source = ColumnDataSource(data=dict(
    est_px=[], est_py=[], est_vx=[], est_vy=[],
    meas_px=[], meas_py=[],
    gt_px=[], gt_py=[], gt_vx=[], gt_vy=[]
))

# Create the figures
radar_fig = figure(title="Radar Measurements", x_axis_label="X", y_axis_label="Y")
lidar_fig = figure(title="Lidar Measurements", x_axis_label="X", y_axis_label="Y")
combined_fig = figure(title="Combined Measurements", x_axis_label="X", y_axis_label="Y")

# Add scatter plots for estimated and ground truth values for each sensor type
radar_fig.circle(x='est_px', y='est_py', source=radar_source, color='red', legend_label='Estimated')
radar_fig.circle(x='meas_px', y='meas_py', source=radar_source, color='blue', legend_label='Measured')
radar_fig.circle(x='gt_px', y='gt_py', source=radar_source, color='green', legend_label='Ground Truth')

lidar_fig.circle(x='est_px', y='est_py', source=lidar_source, color='red', legend_label='Estimated')
lidar_fig.circle(x='meas_px', y='meas_py', source=lidar_source, color='blue', legend_label='Measured')
lidar_fig.circle(x='gt_px', y='gt_py', source=lidar_source, color='green', legend_label='Ground Truth')

combined_fig.circle(x='est_px', y='est_py', source=combined_source, color='red', legend_label='Estimated')
combined_fig.circle(x='meas_px', y='meas_py', source=combined_source, color='blue', legend_label='Measured')
combined_fig.circle(x='gt_px', y='gt_py', source=combined_source, color='green', legend_label='Ground Truth')

# Add hover tools
radar_hover = HoverTool(tooltips=[("Estimate", "@est_px, @est_py"), ("Measured", "@meas_px, @meas_py"), ("Ground Truth", "@gt_px, @gt_py")])
lidar_hover = HoverTool(tooltips=[("Estimate", "@est_px, @est_py"), ("Measured", "@meas_px, @meas_py"), ("Ground Truth", "@gt_px, @gt_py")])
combined_hover = HoverTool(tooltips=[("Estimate", "@est_px, @est_py"), ("Measured", "@meas_px, @meas_py"), ("Ground Truth", "@gt_px, @gt_py")])

radar_fig.add_tools(radar_hover)
lidar_fig.add_tools(lidar_hover)
combined_fig.add_tools(combined_hover)

# Update the data sources with all the points
radar_data = {
    'est_px': [float(values[1]) for values in data if values[0] == 'radar'],
    'est_py': [float(values[2]) for values in data if values[0] == 'radar'],
    'est_vx': [float(values[3]) for values in data if values[0] == 'radar'],
    'est_vy': [float(values[4]) for values in data if values[0] == 'radar'],
    'meas_px': [float(values[5]) for values in data if values[0] == 'radar'],
    'meas_py': [float(values[6]) for values in data if values[0] == 'radar'],
    'gt_px': [float(values[7]) for values in data if values[0] == 'radar'],
    'gt_py': [float(values[8]) for values in data if values[0] == 'radar'],
    'gt_vx': [float(values[9]) for values in data if values[0] == 'radar'],
    'gt_vy': [float(values[10]) for values in data if values[0] == 'radar']
}

lidar_data = {
    'est_px': [float(values[1]) for values in data if values[0] == 'lidar'],
    'est_py': [float(values[2]) for values in data if values[0] == 'lidar'],
    'est_vx': [float(values[3]) for values in data if values[0] == 'lidar'],
    'est_vy': [float(values[4]) for values in data if values[0] == 'lidar'],
    'meas_px': [float(values[5]) for values in data if values[0] == 'lidar'],
    'meas_py': [float(values[6]) for values in data if values[0] == 'lidar'],
    'gt_px': [float(values[7]) for values in data if values[0] == 'lidar'],
    'gt_py': [float(values[8]) for values in data if values[0] == 'lidar'],
    'gt_vx': [float(values[9]) for values in data if values[0] == 'lidar'],
    'gt_vy': [float(values[10]) for values in data if values[0] == 'lidar']
}

radar_source.data = radar_data
lidar_source.data = lidar_data
combined_source.data = {**radar_data, **lidar_data}

# Show the plots
output_notebook()
show(radar_fig)
show(lidar_fig)
show(combined_fig)

```
